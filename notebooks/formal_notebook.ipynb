### Munging

```python
# Time Parsing Function
def time_parse(t_str):
    if not t_str:
        return None
    
    t_str = t_str.split('+')[0]
    
    formats = ['%Y-%m-%dT%H:%M:%S.%f',
               '%Y-%m-%dT%H:%M:%S',
               '%Y-%m-%dT%H:%M:%SZ',
               ]

    for fmt in formats:
        try:
            dt = datetime.datetime.strptime(t_str,fmt)
            return dt
        except:
            continue
        
        
### Copy for reloading without re-parsing
data = raw_data.copy()

# Parse time into datetime object
data['time'] = data['time'].apply(time_parse)
# Date is often useful for binning
data['date'] = data['time'].apply(lambda x: x.date() if pd.notnull(x) else x)

# Add time difference to data; requires a user time series, hence, the "groupby" operation
grouped = data.groupby('username')
data['dt'] = None
for n,g in grouped:
    tseries = g.sort('time')
    data.ix[tseries.index,'dt'] = tseries.time.diff().shift(-1)

data['dt'] = data['dt'].apply(lambda x: x.total_seconds() if pd.notnull(x) else None)
```
